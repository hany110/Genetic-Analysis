The OLS has multiple filters and which can be used interactively, however, for most data collection/gathering/integration tasks automated program scripts will be more useful. OLS does have a tree and graph based data visualization along with a user friendly view for more details which is beneficial for analysis, however, it can be programmed as well.                                                   

```python
import requests
import pprint

# OLS API endpoint for term information
ols_api_url = "https://www.ebi.ac.uk/ols/api/terms"

# Query parameters for the term "Parkinson Disease" in MONDO ontology
params = {
    "iri": "http://purl.obolibrary.org/obo/MONDO_0005180",  # IRI for Parkinson Disease in MONDO
    "obo_id": "MONDO:0005180",  # MONDO ID for Parkinson Disease
    "short_form": "MONDO_0005180",  # Short form ID for Parkinson Disease
    "include": "children",  # Include children of the term
    "size": 1,  # Maximum number of results to return
    "format": "json",  # Format of the response
}

# Make GET request to OLS API
response = requests.get(ols_api_url, params=params)

# Check if request was successful
if response.status_code == 200:
    # Parse JSON response
    data = response.json()

    pprint.pprint(data)
```

```python
children_link = data["_embedded"]["terms"][0]["_links"]["children"]["href"]
print(children_link)
```

```python
response = requests.get(children_link)

# Check if request was successful
if response.status_code == 200:
    # Parse JSON response
    data_c = response.json()

    # Extract direct children from the response
    # children = data['_embedded']['terms'][0]['_embedded']['children']
    pprint.pprint(data_c)
```

```python
# Extracting label and obo_id for each element in the "terms" array
for term in data_c["_embedded"]["terms"]:
    label = term["label"]
    obo_id = term["obo_id"]
    print("Name:", label)
    print("Code:", obo_id)
    print()
```
